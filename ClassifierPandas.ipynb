```python
def match_player_attributes(match_instance, player_attributes, players):
    match_id =  match_instance.match_api_id
    date = match_instance['date']
    
    player_attr_instance = pd.DataFrame() #new df
    player_rating =[]
    for player in players:
        player_id = match_instance[player]
        attr = player_attributes[player_attributes.player_api_id == player_id]
    
        attr_tilldate = attr[attr.date < date].sort_values(by = 'date', ascending = False)[:1]
        attr_tilldate.reset_index(inplace = True, drop = True)
        overall_rating = pd.Series(attr_tilldate.loc[0, "overall_rating"])
    
        rating = "{}_overall_rating".format(player)
        player_rating.append(rating)
        player_attr_instance = pd.concat([player_attr_instance, overall_rating], axis = 1)

    
    player_attr_instance.columns = player_rating        
    player_attr_instance['match_api_id'] = match_id

    player_attr_instance.reset_index(inplace = True, drop = True)
    return player_attr_instance.loc[0]
    
```

```python
players = ['home_player_1', 'home_player_2', 'home_player_3', "home_player_4", "home_player_5",
            "home_player_6", "home_player_7", "home_player_8", "home_player_9", "home_player_10",
            "home_player_11", "away_player_1", "away_player_2", "away_player_3", "away_player_4",
            "away_player_5", "away_player_6", "away_player_7", "away_player_8", "away_player_9",
            "away_player_10", "away_player_11"]

def merge_player_attributes(match_table, player_attributes_table):
    merged_tables = match_table.apply(lambda x: match_player_attributes(x, player_attributes, players), axis = 1)
    return merged_tables

xy = merge_player_attributes(match, player_attributes)
xy
```

```python
from sklearn import svm
from sklearn.model_selection import train_test_split
features = ['country_id', 'league_id', 'stage', 'match_api_id',
       'home_team_api_id', 'away_team_api_id', 'home_team_goal',
       'away_team_goal','home_player_1_overall_rating',
       'home_player_2_overall_rating', 'home_player_3_overall_rating',
       'home_player_4_overall_rating', 'home_player_5_overall_rating',
       'home_player_6_overall_rating', 'home_player_7_overall_rating',
       'home_player_8_overall_rating', 'home_player_9_overall_rating',
       'home_player_10_overall_rating', 'home_player_11_overall_rating',
       'away_player_1_overall_rating', 'away_player_2_overall_rating',
       'away_player_3_overall_rating', 'away_player_4_overall_rating',
       'away_player_5_overall_rating', 'away_player_6_overall_rating',
       'away_player_7_overall_rating', 'away_player_8_overall_rating',
       'away_player_9_overall_rating', 'away_player_10_overall_rating',
       'away_player_11_overall_rating']
label = ['Winner2']
df_X = joined_df[features]
df_Y = joined_df[label]

X_train, X_test, Y_train, Y_test = train_test_split(df_X, df_Y, test_size=0.33, random_state=42)
```
